```python
import matplotlib. pyplot as plt
import numpy as np
import math
from scipy.integrate import odeint
```

```python
def fun(n,r,x_0):
    x_f = 0
    for i in range(n):
        x_f = r*x_0*(1-x_0)
        if x_f < 1:
            x_0 = x_f
        if x_f > 1:
            x_0 = x_f%1
    return x_0
```

```python
n=50
r=3.9
F = [[] for x in range(5)]
for i in range(len(F)):
    x_0 = float(input('Valor de x inicial = '))
    for j in range(n):
        F[i].append(fun(j,r,x_0))
```

```python
plt.figure(figsize=(20,5))
for i in range(len(F)):
    plt.plot(np.linspace(1,n,n),F[i],'o-')
```

```python
n=50
r=np.linspace(2,4,1000)
x_0 = np.linspace(0,1,1000)
F = [[] for x in range(len(x_0))]
for i in range(len(x_0)):
    for j in range(len(r)):
        F[i].append(fun(n,r[j],x_0[i]))
```

```python
plt.figure(figsize=(15,15))
for i in range(len(F)):
    plt.xlabel('$A$',fontsize=18)
    plt.ylabel('$x_n$',fontsize=18 )
    plt.plot(r,F[i],'k,')
plt.savefig("map_log.jpg")
```
