# Mapeo Logístico 

Definimos las librerias:

```python
import matplotlib. pyplot as plt
import numpy as np
import math
from scipy.integrate import odeint
```

El mapeo logístico:

\begin{equation}
x_{n+1} = Ax_{n} (1-x_{n})
\end{equation}

es un mapeo iterativo donde x toma valores entre 0 y 1 y A es un valor cualquiera.

Definimos una función para calcular el valor $x_{n+1}$ tal que $0<x_{n+1}<1$

```python
def fun(n, A, x_0):
    x_f = 0 
    for i in range(n): #ciclo para iterar el mapeo n veces
        x_f = A*x_0*(1-x_0) 
        x_0 = x_f
    return x_0
```

Así, para $A = 0.2$ y $10$ años despues, podemos ver como va evolucionando la población para diferentes poblaciones con $x$ iniciales distintas:

```python
n_1 = 10 #número de años que pasan 
A_1 = 0.2 #parametro A

F_1 = [[] for x in range(5)]
for i in range(len(F_1)):
    x_0 = float(input('Valor de x inicial = '))
    for j in range(n_1):
        F_1[i].append(fun(j,A_1,x_0))
```

```python
#Graficamos
plt.figure(figsize=(20,5))
for i in range(len(F_1)):
    plt.plot(np.linspace(1,n_1,n_1),F_1[i],'o-')
plt.xlabel('Población', fontsize = 20)
plt.ylabel('Años', fontsize = 20)
```

Vemos como cada población se extingue al pasar 4 años 

Si cambiamos el parametros $A$ a un valor mayor como $A = 1.6$ y la misma población inicial

```python
n_2 = 15 #número de años que pasan 
A_2 = 1.6 #parametro A

F_2 = [[] for x in range(5)]
for i in range(len(F_2)):
    x_0 = float(input('Valor de x inicial = '))
    for j in range(n_2):
        F_2[i].append(fun(j,A_2,x_0))
```

```python
#Graficamos
plt.figure(figsize=(20,5))
for i in range(len(F_2)):
    plt.plot(np.linspace(1,n_2,n_2),F_2[i],'o-')
plt.xlabel('Población', fontsize = 20)
plt.ylabel('Años', fontsize = 20)
```

Vemos como algunas poblaciones evolucionan de manera distinta al pasar de los años pero todas llegan a tener la misma cantida de población $\approx 0.4$ al pasar 9 años 

Pasa algo interesante si tomamos valores $2<A<4$, por ejemplo si tomamos $A = 3.3$

```python
n_3 = 30 #número de años que pasan 
A_3 = 3.7 #parametro A

F_3 = [[] for x in range(5)]
for i in range(len(F_3)):
    x_0 = float(input('Valor de x inicial = '))
    for j in range(n_3):
        F_3[i].append(fun(j,A_3,x_0))
```

```python
#Graficamos
plt.figure(figsize=(20,5))
for i in range(len(F_3)):
    plt.plot(np.linspace(1,n_3,n_3),F_3[i],'o-')
plt.xlabel('Población', fontsize = 20)
plt.ylabel('Años', fontsize = 20)
```

Aquí la evolución de cada población es distinta a las demas al pasar de ciarta cantidad de años $\approx 20$ la cantidad de pobración tiende a tomar 2 números espesificos $0.5$ y $0.8$. 

mostramos todas las graficas anteriores para compararlas:

```python
plt.figure(figsize=(20,20))
plt.title('')
plt.subplot(3,1,1)
for i in range(len(F_1)):
    plt.plot(np.linspace(1,n_1,n_1),F_1[i],'ko-')
plt.ylabel('Población (A = 0.2)', fontsize = 25)
plt.title('Evolución de pobraciones',fontsize = 30)

plt.subplot(3,1,2)
for i in range(len(F_2)):
    plt.plot(np.linspace(1,n_2,n_2),F_2[i],'ko-')
plt.ylabel('Población (A = 1.6)', fontsize = 25)

plt.subplot(3,1,3)
for i in range(len(F_3)):
    plt.plot(np.linspace(1,n_3,n_3),F_3[i],'ko-')
plt.xlabel('Años', fontsize = 25)
plt.ylabel('Población (A = 3.7)', fontsize = 25)
plt.savefig("logistico.jpg") #guardamos la gráfica como imagen jpg
```

Al hacer un diagrama de bifurcación del número $x_n$ de pobladores al pasar una cantidad consederable de años tendremos lo siguiente: 

```python
n = 300
r = np.linspace(2,4,1000)
x_0 = np.linspace(0,1,1000)
F = [[] for x in range(len(x_0))]
for i in range(len(x_0)):
    for j in range(len(r)):
        F[i].append(fun(n,r[j],x_0[i]))
```

```python
plt.figure(figsize=(15,15))
for i in range(len(F)):
    plt.xlabel('$A$',fontsize=25)
    plt.ylabel('$x_n$',fontsize=25 )
    plt.plot(r,F[i],'k,')
plt.title('Diagrama de bifurcaión del mapeo logístico',fontsize = 40)
plt.savefig("map_log.jpg")
```
