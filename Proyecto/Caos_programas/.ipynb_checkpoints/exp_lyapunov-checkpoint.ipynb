# Exponentes de Lyapunov

```python
import matplotlib. pyplot as plt
import numpy as np
import math
import random 
from scipy.integrate import odeint
```

```python
def lyap_coef (n, A):
    ly = []
    l = 0
    for i in range(len(A)): #ciclo para iterar el mapeo n veces
        x_f = 0 
        l = 0
        x_0 = random.random()
        for j in range(n):
            x_f = A[i]*x_0*(1-x_0) 
            x_0 = x_f
            l = l + math.log(abs(A[i]-2*A[i]*x_f))
        ly.append(l/n) 
    return ly
```

```python
n = 300 #Número de años que pasan 
A = np.linspace(3,4,10000) 
```

```python
lyp = lyap_coef(n,A)
```

```python
plt.figure(figsize=(15,15))
plt.xlabel('$A$',fontsize=35)
plt.ylabel('$x_n$',fontsize=35 )
plt.plot(A,lyp,'k-')
plt.title('Coeficientes de Lyapunov', fontsize = 40)
plt.savefig("lyapunov.jpg")
```
