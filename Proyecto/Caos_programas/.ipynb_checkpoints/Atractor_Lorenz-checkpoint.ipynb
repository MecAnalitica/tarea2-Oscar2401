# Atractor de Lorenz

```python
import matplotlib. pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import math
from scipy.integrate import odeint
```

\begin{equation}
\frac{dx}{dt} = a(y-x)
\end{equation}

\begin{equation}
\frac{dy}{dt} = x(b-z)-y
\end{equation}

\begin{equation}
\frac{dz}{dt} = xy-cz
\end{equation}

```python
def sol_atrac(var_ini, t,a,b,c ):
    x,y,z = var_ini
    return [a*(y-x), x*(b-z)-y, x*y-c*z]
```

```python
x = 1
y = 1
z = 1
a = 10
b = 28 #<--- Variamos este
c = 8/3
t = np.linspace(0, 100, 100000)
var_ini=[x,y,z]
atrac =odeint(sol_atrac,var_ini,t, args=(a,b,c))
```

```python
plt.subplots(1, figsize=(12, 12),subplot_kw={'projection':'3d'})
plt.plot(atrac[:,0], atrac[:,1], atrac[:,2], 'k,')
plt.title('Atractor de Lorenz', fontsize=20)
plt.xlabel('$x$', fontsize=25)
plt.ylabel('$y$', fontsize=25)
plt.savefig("atrac_lor.jpg")
```
