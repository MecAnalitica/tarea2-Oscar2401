# Ejemplo del oscilador armonico forzado

```python
import matplotlib. pyplot as plt
import numpy as np
import math
from scipy.integrate import odeint
```

Ecuaci√≥n a resolver:
\begin{equation}
\ddot \theta + \mu \dot\theta + sin(\theta)  = Acos(\omega_d \theta) 
\end{equation}

```python
def fun_com(con_ini,t,mu,omg_0,A,omg_d):
    theta, omega, phi = con_ini 
    return [omega, -mu*omega - pow(omg_0,2)*np.sin(theta) + A * np.cos(phi),omg_d]
```

```python
t = np.arange(0, 5.0, 0.001)
#condiciones iniciale y parametros:
mu = 3*np.pi/2
omg_0 = 3*np.pi
omg_d = 2*np.pi
th_i = -0.1
dth_i = 2
phi = 0
con_ini =[th_i,dth_i,phi]
sol_4a = []
A = 5
sol_4a = (odeint(fun_com,con_ini,t,args = (mu,omg_0,A,omg_d)))
```

```python
plt.figure(figsize=(10,10))
plt.title('Espacio fase',fontsize = 25, color = 'k', verticalalignment = 'baseline', horizontalalignment = 'center')
plt.plot(sol_4a[:, 0],sol_4a[:, 1], 'k-')
plt.xlabel('$\Theta$', fontsize = 25)
plt.ylabel('$\dot\Theta$',fontsize = 25)
plt.grid()
plt.savefig("pendulo.jpg")
```
