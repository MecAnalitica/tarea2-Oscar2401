# Coeficientes de Lyapunov

```python
import matplotlib. pyplot as plt
import numpy as np
import math
import random 
from scipy.integrate import odeint
```

Se encontraran los coeficientes $\lambda$ de Lyapunov para el mapero logístico ($x_{n+1} = f_A(x_n) = Ax_n(1-x_n)$)

Definimos la función
\begin{equation}
\lambda = \frac{1}{N} \sum_{i = 1}^N \lambda(x_i)
\end{equation}

```python
def lyap_coef (n, A):
    ly = [] 
    for i in range(len(A)): #ciclo para encontrar lambda de cada mapeo con distinta A
        x_f = 0 
        l = 0 # acumulador de la suma de los coeficientes 
        x_0 = random.random() #x inicial aleatoria entre 0 y 1
        for j in range(n):
            x_f = A[i]*x_0*(1-x_0) #encontramos el valor de x_i 
            x_0 = x_f
            l = l + math.log(abs(A[i]-2*A[i]*x_f))  #derivada del mapeo logístico
        ly.append(l/n) #se agrega el coeficiente del mapeo A[i]
    return ly #La función regresa un arreglo con los coeficientes para cada mapeo con A distinta
```

```python
n = 800 #Número de años que pasan 
A = np.linspace(3,4,10000) #valores del coeficiente A entre 3 y 4 
```

```python
lyp = lyap_coef(n,A) #calculamos los coeficiente de Lyapunov
```

```python
#Graficamos
plt.figure(figsize=(8,8))
plt.xlabel('$A$',fontsize=40)
plt.ylabel('$\lambda$',fontsize=35 )
plt.plot(A,lyp,'k-')
plt.title('Coeficientes de Lyapunov', fontsize = 20)
plt.savefig("lyapunov.jpg")
```
