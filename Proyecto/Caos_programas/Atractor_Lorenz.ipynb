# Atractor de Lorenz

Definimos las librerias:

```python
import matplotlib. pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import math
from scipy.integrate import odeint
```

Las ecuaciones del modelo de Lorenz son las siguientes:

\begin{equation}
\frac{dx}{dt} = a(y-x)
\end{equation}

\begin{equation}
\frac{dy}{dt} = x(b-z)-y
\end{equation}

\begin{equation}
\frac{dz}{dt} = xy-cz
\end{equation}

las caules las definimos en la siguinete funci√≥n:

```python
def sol_atrac (var_ini, t, a, b, c ):
    x,y,z = var_ini #Valores iniciales
    return [a*(y-x), x*(b-z)-y, x*y-c*z]
```

Con ayuda de la libreria scipy resolvemos las ecuaciones anteriores con condiciones iniciales:

\begin{equation}
\begin{split}
x_0 &= 10 \\
y_0 &= -30 \\
z_0 &= -5 \\
\end{split}
\end{equation}

y parametros:

\begin{equation}
\begin{split}
a &= 10 \\
b &= 28 \\
c &= \frac{8}{3} \\
\end{split}
\end{equation}

```python
#Definimos las condiciones iniciales (C.I.)
x = 10
y = -30
z = -5
var_ini = [x,y,z] #guardamos en un arreglo las C.I.

#Definimos los parametros. 
a = 10
b = 28
c = 8/3

t = np.linspace(0, 100, 100000) 

atrac = []
atrac = odeint(sol_atrac,var_ini,t, args=(a,b,c)) #guaramos en un arreglo 'atrac' las soluciones.
```

```python
#Graficamos en 3-Dimenciones 
plt.subplots(1, figsize=(12, 12),subplot_kw={'projection':'3d'})
plt.plot(atrac[:,0], atrac[:,1], atrac[:,2], 'k-')
plt.title('Atractor de Lorenz', fontsize=40)
plt.xlabel('$x$', fontsize=35)
plt.ylabel('$y$', fontsize=35)
plt.savefig("atrac_lor.jpg") #guardamos la gr√°fica como imagen jpg
```
