# Ecuaciones Lotka–Volterra

```python
import matplotlib. pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import math
from scipy.integrate import odeint
```

\begin{equation}
\begin{split}
\frac{dx}{dt} &= x(A-By)\\
\frac{dy}{dt} &= -y(C-Dx)
\end{split}
\end{equation}

```python
def lok_vol (var_ini, t,par_r, par_1,par_2,par_3,par_4, s ):
    x,y,z,w = var_ini #Valores iniciales
    r1,r2,r3,r4 = par_r
    a11,a12,a13,a14 = par_1 
    a21,a22,a23,a24 = par_2 
    return [ r1*x*(1 - a11*x - s*a12*y - s*a13*z - s*a14*w),r2*y*(1 - s*a21*x - a22*y - s*a23*z - s*a24*w)
```

```python
#Definimos las condiciones iniciales (C.I.)
x = 1
y = 1
var_ini = [x,y] #guardamos en un arreglo las C.I.

#Definimos los parametros. 
s = np.linspace(0.8, 1.4, 100)
a = 0.3013
b = 0.4586
c = 0.1307
d = 0.3343

t = np.linspace(0, 100, 100) 

F = [[] for x in range(len(s))]
for i in range(len(s)):
    for j in range(len(s)):
        F[i].append(odeint(lok_vol,var_ini,t, args=(a,b,c,d,s[j]))[:,0])
```

```python
plt.figure(figsize=(15,15))
for i in range(len(s)):
    plt.xlabel('$A$',fontsize=25)
    plt.ylabel('$x_n$',fontsize=25 )
    plt.plot(s,F[i][len(F)-1],'k-')
plt.title('Diagrama de bifurcaión del mapeo logístico',fontsize = 40)
```

```python
def lok_vol (var_ini, t,par_r, par_1,par_2,par_3,par_4, s ):
    x,y,z,w = var_ini #Valores iniciales
    r1,r2,r3,r4 = par_r
    a11,a12,a13,a14 = par_1 
    a21,a22,a23,a24 = par_2 
    a31,a32,a33,a34 = par_3
    a41,a42,a43,a44 = par_4
    return [ r1*x*(1 - a11*x - s*a12*y - s*a13*z - s*a14*w),r2*y*(1 - s*a21*x - a22*y - s*a23*z - s*a24*w),r3*z*(1 - s*a31*x - s*a32*y - a33*z - s*a34*w ), r4*w*(1 - s*a41*x - s*a42*y - s*a43*z - a44*w) ]
```

```python
#Definimos las condiciones iniciales (C.I.)
x = 1
y = 1
z = 1
w = 1
var_ini = [x,y,z,w] #guardamos en un arreglo las C.I.

#Definimos los parametros. 
par_r = [1,0.72,1.53,1.27]
par_1 = [1.00,1.09,1.52,0.00]
par_2 = [0.00,1.00,0.44,1.36]
par_3 = [2.33,0.00,1.00,0.47]
par_4 = [1.21,0.51,0.35,1.00]

s = 2.074
t = np.linspace(0, 100, 1000000) 

lotka = []
lotka = odeint(lok_vol,var_ini,t, args=(par_r,par_1,par_2,par_3,par_4,s)) #guaramos en un arreglo 'atrac' las soluciones.
```

```python
#Graficamos en 3-Dimenciones 
plt.subplots(1, figsize=(12, 12),subplot_kw={'projection':'3d'})
plt.plot(lotka[:,1], lotka[:,2], lotka[:,3], 'k-')
plt.title('Atractor de Lorenz', fontsize=40)
plt.xlabel('$x$', fontsize=35)
plt.ylabel('$y$', fontsize=35)
plt.savefig("lotka.jpg") #guardamos la gráfica como imagen jpg
```
