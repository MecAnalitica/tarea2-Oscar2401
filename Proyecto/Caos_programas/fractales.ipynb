### Triángulo de Sierpinski

Usando el triángulo de pascal:



```python
a, b = [], [] #iniciamos las variables

R = 1024 #cantidad de puntos 
for j in range(R):
    n = j
    k = 0
    for i in range(n):
        k = i
        c = (math.factorial(n))//((math.factorial(k)*(math.factorial(n-k)))) #convinaciones 
        if c%2 == 0: # si la convinacion es par se guarda en el arreglo
            a.append(i)
            b.append(j)
            
```

```python
plt.figure(figsize=(20,20))
plt.title('Triangulo de Sierpinski',fontsize=30)
plt.plot(a,b,'k,')
plt.savefig("triangulo.jpg")
```

### Conjuntos de julia

Función compleja iterativa:
\begin{equation}
z_0 = z
\end{equation}
\begin{equation}
z_{n+1} = z_{n}^2 +c
\end{equation}

donde $|z_n| >2$ la sucesión diverge y el punto $z$ no pertenece al conjunto de Julia. 

```python
iters = 800 
n_p = 800
x = np.linspace(-2.25,0.75,n_p).astype(np.float32)
y = np.linspace(-1.25,1.25,n_p).astype(np.float32)
N = np.zeros([n_p,n_p],int)
for u,a in enumerate(x):
    for v,b in enumerate(y):
        z = complex(0,0) #z inicial 
        c = complex(a,b) # constante c
        for i in range(iters):
            z = z**2 + c
            if abs(z) > 2: #restricción
                N[v,u] = 1
                break
```

```python
fig = plt.figure(figsize=(10, 10), dpi=72)
light = colors.LightSource(azdeg=315, altdeg=10)
M = light.shade(N, cmap=plt.cm.hot, vert_exag=1.5, norm=colors.PowerNorm(0.3), blend_mode='hsv')
plt.imshow(M)
plt.savefig("fractal.jpg")
plt.show()
```
